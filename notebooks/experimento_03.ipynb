# Teste de execução com arquivo único e leitura de csv

```python
import pandas as pd
import sys
import os

# 1. Configuração de Caminhos
sys.path.append(os.path.abspath(os.path.join('..')))
from src.components.transformersCasoNome import processar_usuarios_etl

# Definindo local de entrada e saída
caminho_entrada = os.path.join('..', 'data', 'vendas.csv')
caminho_saida = os.path.join('..', 'data', 'vendas_processadas.csv')

# 2. EXTRAÇÃO (Extract)
try:
    df_bruto = pd.read_csv(caminho_entrada)
    print(f"✔️ Leitura concluída: {len(df_bruto)} registros encontrados.")
except FileNotFoundError:
    print(f"❌ Erro: O arquivo {caminho_entrada} não existe.")
    df_bruto = pd.DataFrame()

# 3. TRANSFORMAÇÃO E CARGA (Transform & Load)
if not df_bruto.empty:
    # Aplicando a lógica de nomes brasileiros e preposições que criamos
    print("⚙️ Processando nomes brasileiros e aplicando regras de negócio...")
    df_transformado = processar_usuarios_etl(df_bruto, estrategia='primeiro_termo')
    
    # Gerando o novo arquivo CSV
    # 'utf-8-sig' garante que o Excel no Windows abra com acentos corretos
    df_transformado.to_csv(caminho_saida, index=False, encoding='utf-8-sig')
    
    print(f"✔️ Sucesso! O arquivo tratado foi salvo em: {caminho_saida}")
    
    # Exibe uma prévia do resultado para o professor conferir em sala
    print("\n--- PRÉVIA DOS DADOS TRATADOS ---")
    display(df_transformado.head())
```
